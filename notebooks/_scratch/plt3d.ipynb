```python
from cuber import Cube
olls = {}
sune = "RUR'URU2R'"
antisune = "L'U'LU'L'U2L"
c = Cube(3, whiteplastic=True)
c.draw_top = True
c.echo = False

c.reset()
c.move('x2')
c.reverse(sune)
olls['fish'] = c.copy()

c.reset()
c.move('x2')
c.reverse(sune)
c.reverse(sune)
olls['double-headlights'] = c.copy()

c.reverse(sune)
c.move('U2')
olls['stupid-fish'] = c.copy()

c.reset()
c.move('x2')
c.reverse(sune)
c.move('U2')
c.move(sune)
olls['headlights'] = c.copy()


c.reset()
c.move('x2')
c.reverse(antisune)
c.move(sune)
olls['chamelon-headlights'] = c.copy()


c.reset()
c.move('x2')
c.reverse(antisune)
c.move(sune)
c.reverse(antisune)
c.move("U")
olls['chameleon'] = c.copy()
```

```python
from IPython.display import Markdown
for key, c in olls.items():
    c = c.copy()
    c.echo = False
    c.draw_front = False
    c.top_solving = 'oll'
    display(Markdown(f"# {key}"))
    display(c)
    display(Markdown(f"### ==> Move: sune = {sune}"))
    c.move(sune)
    display(c)
```
